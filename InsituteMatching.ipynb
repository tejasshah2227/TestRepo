```python
import pandas as pd
import re
import numbers
import numpy as np
import mpu
import math
import geopy.distance
import os
```

### Importing txt files and storing it in Pandas Dataframe

```python
file1=pd.read_csv('inst_list_geocodes.txt',delimiter='\t')
file1.columns=['Name','Country', 'Latitude', 'Longitude']
file1.head()
```

```python
file2=pd.read_csv("inst_mast_list_geocodes.txt",delimiter="\t")
file2.columns=['Name', 'City', 'Country', 'Latitude', 'Longitude']
file2.tail(10)
```

### Checking results for the first 50 records:

```python
file1=file1.head(50)
```

### Distance logic:

```python
#For minimum distance
min_dist=0
i=0

os.remove('output1.txt') if os.path.exists('output1.txt') else None

#Traversing each row in first txt file
for row1 in file1.itertuples():
    
    #Store the latitude and longitude for further computations:
    lat1=row1.Latitude
    long1=row1.Longitude
    
    for row2 in file2.itertuples():
        i=i+1
        
        #Store latitude and longitude o each row to perform computation one by one.
        lat2=row2.Latitude
        long2=row2.Longitude
        
        
        #If any of the values ie (lat1,long1) and (lat2,long2) is Null, it will not perform operation.
        if(math.isnan(lat1)==False and math.isnan(long1)==False and math.isnan(lat2)==False and math.isnan(lat2)==False):
            #dist=geopy.distance.vincenty((lat1,long1), (lat2,long2)).km
            
            #Method calculates haversine_distance given latitudes and longitudes
            #https://en.wikipedia.org/wiki/Haversine_formula
            #https://mpu.readthedocs.io/en/latest/mpu.html
            
            dist = mpu.haversine_distance((lat1, long1), (lat2, long2))
            
            
            #print('Distance in kms',dist)
            #dlist.append(dist)
            
            
            #Assume the first distance as minimum for comparsion
            if(i==1):
                print('\n\n########################Next Record from File1:#############################')
                print('i:',i)
                min_dist=dist
                print('min:',min_dist)
        
            # If distance is smaller than previous min_distance, update the current distance as minimum.
            if(min_dist>dist):
                min_dist=dist
                print('\nUniversity:',row2.Name,' Co-ordinates:(',lat2,',',long2,')','\nNew Min dist:',min_dist)
                
                min_dist=round(min_dist,2)
                
                #Update the information of the corresponding record to be written into output file
                univ_name=str(row2.Name)
                country=str(row2.Country)
                city=str(row2.City)
                res_lat=str(row2.Latitude)
                res_long=str(row2.Longitude)
    i=0
    
    
    #Print final minimum result
    print('\n*********Minimum distance from University:',row1.Name,'with Country:',row1.Country,'and co-ordinates:(',lat1,',',long1,') is: ',round(min_dist,2),'\n Selected University:',
         univ_name,' Country:', country,' Co-ordinates:(',lat2,',',long2,')')
    
    #Write it in txt file
   
    
    with open('output1.txt', 'a',encoding="utf-8") as f:
        f.write(str(row1.Name) + "\t" + str(row1.Country) + "\t" + str(row1.Latitude) + "\t" + str(row1.Longitude) + "\t" 
                + str(univ_name) + "\t" + str(country) + "\t" + str(res_lat) + "\t" + str(res_long) + "\t" +str(city)
                +'\t'+str(min_dist)+"\n\n")
    #print('Minimum distance of:',row1.Name ,'with coordinates:',lat1,',',long1,'is: ',min(dlist))
#dlist=[]
```
